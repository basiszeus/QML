```python
import numpy as np
from qutip import Bloch, Qobj
```

```python
# General qubit state: |ψ⟩ = α|0⟩ + β|1⟩
alpha = 1/np.sqrt(2)
beta = 1/np.sqrt(2)   # example: |+> state
```

```python
# Initial state
psi_init = Qobj([alpha, beta])

# Define Phase gates
phi_S = np.pi/2   # S gate
phi_T = np.pi/4   # T gate
```

```python
P_S = np.array([[1, 0],
                [0, np.exp(1j*phi_S)]])

P_T = np.array([[1, 0],
                [0, np.exp(1j*phi_T)]])
```

```python
# Apply Phase gates
result_S = P_S @ np.array([alpha, beta])
result_T = P_T @ np.array([alpha, beta])

psi_S = Qobj(result_S)
psi_T = Qobj(result_T)
```

```python
# Bloch sphere for S gate
b1 = Bloch()
b1.add_states(psi_init)   # before
b1.add_states(psi_S)      # after S
b1.show()

# Bloch sphere for T gate
b2 = Bloch()
b2.add_states(psi_init)   # before
b2.add_states(psi_T)      # after T
b2.show()
```
