```python
import numpy as np
from qutip import Bloch, Qobj
```

```python
# Define Hadamard matrix
H = (1/np.sqrt(2)) * np.array([[1, 1],
                               [1, -1]])

# Basis states
zero = np.array([[1],
                 [0]])   # |0>
one  = np.array([[0],
                 [1]])   # |1>
```

```python
# Apply H to |0>
result_zero = H @ zero
print("H|0> =", result_zero)

# Apply H to |1>
result_one = H @ one
print("H|1> =", result_one)
```

```python
# Apply H to a general state α|0> + β|1>
alpha, beta = 1/np.sqrt(3), np.sqrt(2/3)  # example normalized state
psi = np.array([[alpha],
                [beta]])

result_general = H @ psi
print("H|ψ> =", result_general)
```

---

```python
# Convert to Qobj (column vectors)
plus_q = Qobj(result_zero)
minus_q = Qobj(result_one)

# Bloch sphere
b = Bloch()
b.add_states(plus_q)   # H|0>
b.add_states(minus_q)  # H|1>
b.show()
```
