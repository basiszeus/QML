```python
from qiskit import QuantumCircuit
import numpy as np

# Create a 1-qubit quantum circuit
qc = QuantumCircuit(1)

# Apply Rx rotation of 90 degrees (pi/2)
theta = np.pi / 2
qc.rx(theta, 0)

# Draw the circuit to see what we did
qc.draw('mpl')
```

```python
from qiskit import QuantumCircuit
import numpy as np

qc = QuantumCircuit(1)

# Step 1: Flip |0> -> |1>
qc.x(0)

# Step 2: Apply a rotation around the X axis
qc.rx(np.pi/2, 0)

# Draw the circuit
qc.draw('mpl')
```

```python
from qiskit.quantum_info import Statevector
from qiskit.visualization.bloch import Bloch
import matplotlib.pyplot as plt

# Get the statevector
state = Statevector.from_instruction(qc)

# Extract Bloch coordinates
x = 2 * np.real(np.conj(state.data[0]) * state.data[1])
y = 2 * np.imag(np.conj(state.data[0]) * state.data[1])
z = np.abs(state.data[0])**2 - np.abs(state.data[1])**2

# Plot manually
bloch = Bloch()
bloch.add_vectors([x, y, z])
bloch.show()
```
