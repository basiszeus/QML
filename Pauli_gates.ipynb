```python
import numpy as np
from qutip import Bloch, Qobj
```

```python
# Define Pauli matrices
X = np.array([[0, 1],
              [1, 0]])
Y = np.array([[0, -1j],
              [1j, 0]])
Z = np.array([[1, 0],
              [0, -1]])
```

```python
# Define computational basis states
ket0 = np.array([[1],
                 [0]])
ket1 = np.array([[0],
                 [1]])
```

```python
# General state example (superposition)
alpha = 1/np.sqrt(2)
beta = 1/np.sqrt(2)
psi = np.array([[alpha],
                [beta]])
```

```python
# Helper function to plot original and transformed states
def plot_gate_effect(gate, state, gate_name=""):
    new_state = gate @ state
    b = Bloch()
    b.add_states([Qobj(state), Qobj(new_state)])
    b.vector_color = ['b', 'r']  # blue = original, red = after gate
    b.title = f"{gate_name}: Blue = original, Red = after gate"
    b.show()
    return new_state
```

---

```python
# X on basis states
X0 = X @ ket0
X1 = X @ ket1
print("X|0> =", X0)
print("X|1> =", X1)

# X on general state
Xpsi = X @ psi
print("X|psi> =", Xpsi)
X_psi = plot_gate_effect(X, psi, "Pauli-X")
```

```python
# Z on basis states
Z0 = Z @ ket0
Z1 = Z @ ket1
print("Z|0> =", Z0)
print("Z|1> =", Z1)

# Z on general state
Zpsi = Z @ psi
print("Z|psi> =", Zpsi)

# Plot Bloch sphere: original -> after Z
Z_psi = plot_gate_effect(Z, psi, "Pauli-Z")
```

```python
# Y on basis states
Y0 = Y @ ket0
Y1 = Y @ ket1
print("Y|0> =", Y0)
print("Y|1> =", Y1)

# Y on general state
Ypsi = Y @ psi
print("Y|psi> =", Ypsi)

# Plot Bloch sphere: original -> after Y
Y_psi = plot_gate_effect(Y, psi, "Pauli-Y")
```
